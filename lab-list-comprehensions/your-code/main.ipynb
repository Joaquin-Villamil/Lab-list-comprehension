### 7. Use a list comprehension and the Pandas `read_csv` and `concat` methods to read all CSV files in the */data* directory and combine them into a single data frame. Display the top 10 rows of the resulting data frame.

```python
import os
import pandas as pd
#print(os.name)
#print(os.getcwd())
file_list = [f for f in os.listdir('../data') if f.endswith('.csv')]
data_sets = [pd.read_csv(os.path.join('../data', f)) for f in file_list]
# nrows =9 | seleccionar hasta cuantas filas importar
# usecols=["1","2","4"] | Seleccionar que columnas importar, se importan en orden
data = pd.concat(data_sets, axis=0)
#print(file_list)
#print(data_sets)
data.head(10).transpose().plot(figsize=(10,10), colormap='rainbow', linewidth=2, title='Ejemplo Grafica')
print(data.head(10))
```
